```python
# Load the pandas library
import pandas as pd

# Set the GitHub URL
url = 'https://raw.githubusercontent.com//mawalz05//PLSC381B_Analyzing_Politics_Python//main//Titanic.csv'

# Import the dataframe
df = pd.read_csv(url)

# View the first five obersvations of the dataframe
df.head(5)
```

```python
# Show dimensions
df.shape
```

```python
# Show statistics
df.describe()
```

```python
# Show the whole dataframe
df
```

```python
# Create a Row (An additional passenger)
new_person = pd.Series(['Mooney, Molly', '1st', 75, 'female', 1, 1], 
                       index = ['Name', 'PClass', 'Age', 'Sex', 'Survived', 'SexCode'])

# Append row 
df = df.append(new_person, ignore_index = True)
```

```python
# Let's look at the new addition
df.tail(5)
```

```python
# Slicing data by rows

# Select first row
df.iloc[0]
```

```python
# Select three rows
print(df.iloc[1:4])
```

```python
# Select three rows
print(df.iloc[:4])
```

```python
# Slicing data by columns (with iloc)
print(df.iloc[:,[0,1]])
```

```python
# Same as above using loc instead of iloc
print(df.loc[:, ["Name", "PClass"]])
```

```python
# Show top five where column 'sex' is 'female'
df[df['Sex'] == 'female'].head(5)
```

```python
# Filter rows by Sex and Age
df[(df['Sex'] == 'female') & (df['Age'] >= 65)]
```

```python
# Replacing 'female' in dataframe with 'woman'
print(df['Sex'].replace("female", "Woman").head(2))
```

```python
# Replacing multiple values at the same time
print(df['Sex'].replace(['female', 'male'], ['Woman', 'Man']).head(5))
```

```python
# Renaming multiple columns at once
print(df.rename(columns = {'PClass': 'Passenger Class', 'Sex': 'Gender'}).head(2))
```

```python
# Calculate descriptive statistics of the age variable
print('Maximum:', df['Age'].max())
print('Minimum:', df['Age'].min())
print('Mean:', df['Age'].mean())
print('Sum:', df['Age'].sum())
print('Count:', df['Age'].count())
```

```python
# Select unique values of the Sex variable
print(df['Sex'].unique())
```

```python
# Display unique values along with the number of times they appear
print(df['Sex'].value_counts())
```

```python
# Select missing values in Age, show two rows
print(df[df['Age'].isnull()].head(10))
```

```python
# Delete column in the dataframe
print(df.drop('Age', axis = 1).head(2))
```

```python
# Drop multiple columns in the dataframe
print(df.drop(['Age','Sex'], axis = 1).head(2))
```

```python
# You can also drop column by index
print(df.drop(df.columns[1], axis = 1).head(2))
```

```python
# Keep rows where the variable sex does not equal male
print(df[df['Sex'] != 'male'].head(5))
```

```python
# Keep row matching a unique value
print(df[df['Name'] == 'Allison, Miss Helen Loraine'].head(2))
```

```python
# Keep rows based on index (keep all but the first row)
print(df[df.index != 0].head(2))
```

```python
# Drop duplicates rows from your dataframe
print(df.drop_duplicates().head(2))
```

```python
# Show number of rows that were dropped
print("Number or Rows in the original DataFrame:", len(df))
print("Number of Rows after deduping:", len(df.drop_duplicates())) #No duplicates were dropped
```

```python
#Drop duplicates based on a column
print(df.drop_duplicates(subset = ['Sex']).head(2))
```
