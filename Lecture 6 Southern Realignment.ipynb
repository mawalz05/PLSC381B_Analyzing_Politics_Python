```python
import pandas as pd
url = r'https://raw.githubusercontent.com/mawalz05/PLSC381B_Analyzing_Politics_Python/main/PartyScores.csv'
df = pd.read_csv(url)

print(df.shape)
df.head(5)
```

```python
keep = ['state_name', 'year', 'party', 'party_strength']
df2 = df[keep]
df2.head(5)
```

```python
state_means = df2.groupby(['state_name', 'party'])['party_strength'].mean().to_frame().reset_index()
state_means
```

```python
Dems = df2[df2['party'] == 'Dem']
Reps = df2[df2['party'] == 'Rep']
```

```python
Dems.sort_values('party_strength', ascending = False)
```

```python
Reps.sort_values('party_strength', ascending = False)
```

```python
Alabama = df2[df2['state_name'] == 'Alabama']
```

```python
Alabama.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.relplot(x = 'year', y = 'party_strength', kind = 'line', hue = 'party', data = Alabama, markers = True, dashes = False)
plt.show()

                                                                   
```

```python
Dems = df[df['party'] == 'Dem']
Reps = df[df['party'] == 'Rep']

Dems_region = Dems.groupby(['region', 'year',])['party_strength'].mean().to_frame().reset_index()
Reps_region = Reps.groupby(['region', 'year',])['party_strength'].mean().to_frame().reset_index()

Dems_region_filtered = Dems_region[(Dems_region['region'] == 1) | (Dems_region['region'] == 5)]
```

```python
sns.relplot(x = 'year', y = 'party_strength', kind = 'line', hue = 'region',
            data = Dems_region_filtered, markers = True, dashes = False)
plt.show()
```
