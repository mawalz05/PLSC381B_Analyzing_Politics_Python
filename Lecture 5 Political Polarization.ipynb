```python
import pandas as pd

url = 'https://raw.githubusercontent.com//mawalz05//PLSC381B_Analyzing_Politics_Python//main//NominateScores.csv'

df = pd.read_csv(url)

print(df.shape)
df.head()
```

```python
# Making data sets for individual congresses
df_115 = df[df['congress'] == 115]
df_110 = df[df['congress'] == 110]
df_105 = df[df['congress'] == 105]
df_100 = df[df['congress'] == 100]
df_95 = df[df['congress'] == 95]
df_90 = df[df['congress'] == 90]
df_85 = df[df['congress'] == 85]
df_80 = df[df['congress'] == 80]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots() #Creating the figure and axis objects
ax.scatter(df_80['nominate_dim1'], df_80['nominate_dim2'], c = df_80['party_code']) #plotting nominate scores by party_code
plt.show()
```

```python
fig, ax = plt.subplots() #Creating the figure and axis objects
ax.scatter(df_95['nominate_dim1'], df_95['nominate_dim2'], c = df_95['party_code']) #plotting nominate scores by party_code
plt.show()
```

```python
fig, ax = plt.subplots() #Creating the figure and axis objects
ax.scatter(df_105['nominate_dim1'], df_105['nominate_dim2'], c = df_105['party_code']) #plotting nominate scores by party_code
plt.show()
```

```python
fig, ax = plt.subplots() #Creating the figure and axis objects
ax.scatter(df_115['nominate_dim1'], df_115['nominate_dim2'], c = df_115['party_code']) #plotting nominate scores by party_code
plt.show()
```

```python
# Let's look at NY
df_ny = df[df['state_icpsr'] == 13]
df_ny.head()
```

```python
fig, ax = plt.subplots() #Creating the figure and axis objects
ax.scatter(df_ny['congress'], df_ny['nominate_dim1'], c = df_ny['party_code']) #plotting nominate scores by party_code
plt.show()
```

```python
# https://queirozf.com/entries/pandas-dataframe-groupby-examples
df_ny2 = df_ny.groupby(['congress', 'party_code'])['nominate_dim1'].mean().to_frame().reset_index()
df_ny2.head()
```

```python
fig, ax = plt.subplots() #Creating the figure and axis objects
ax.scatter(df_ny2['congress'], df_ny2['nominate_dim1'], c = df_ny2['party_code']) #plotting nominate scores by party_code
plt.show()
```

```python
print(df_ny2['party_code'].unique())
```

```python
df_ny2.tail(5)
```

```python
reps = df_ny2[df_ny2['party_code'] == 200]
dems = df_ny2[df_ny2['party_code'] == 100]

fig,ax = plt.subplots()
ax.scatter(reps['congress'], reps['nominate_dim1'], color = 'red', label = "Republicans")
ax.scatter(dems['congress'], dems['nominate_dim1'], color = 'blue', label = "Democrats")
ax.legend()
ax.set_xlabel("Congress")
ax.set_ylabel("Nominate Scores")
plt.show()
```

```python
df2 = df.groupby(['congress', 'party_code'])['nominate_dim1'].mean().to_frame().reset_index()
reps = df2[df2['party_code'] == 200]
dems = df2[df2['party_code'] == 100]

fig,ax = plt.subplots()
ax.scatter(reps['congress'], reps['nominate_dim1'], color = 'red', label = "Republicans")
ax.scatter(dems['congress'], dems['nominate_dim1'], color = 'blue', label = "Democrats")
ax.legend()
ax.set_xlabel("Congress")
ax.set_ylabel("Nominate Scores")
plt.show()
```
